## What is the difference of evaluation methods for Pansharpening?
The most widely used quality measures, namely,
1. the spectral angle mapper (SAM), a spectral measure;
1. the peak signal-to-noise ratio (PSNR);
1. the spatial correlation coefficient (SCC), a spatial measure;
1. the universal image quality index (UIQI or Q-index);
1. the structural similarity (SSIM);
1. the erreur relative globale adimensionnelle de synthese (ERGAS).
1. No reference method, QNR

Denote ground truth by $I\in\mathbb{R}^{HW\times C}$ and generated one by $\hat{I}\in\mathbb{R}^{HW\times C}$,
1. SAM. $\hat{I}_i, I_i \in\mathbb{R}^{C\times 1}$ stand for the $i$-th row of $\hat{I}$ and $I$ respectively,
then
$$
\mathrm{SAM}(\hat{I}, I) = \frac{1}{HW}\sum_{i=1}^{HW}\arccos \frac{<\hat{I}_i, I_i>}{||\hat{I}_i||||I_i||} \,.
$$

1. PSNR. $\mathrm{MAX}$ is the dynamic range of the pixel-values (typically, $2^8 - 1$), then
$$
\mathrm{PSNR}(\hat{I}, I) = 10\log_{10} \left(\frac{\mathrm{MAX}}{\mathrm{RMSE}(\hat{I}, I)}\right)^2\,.
$$

1. SCC. $\hat{I}_c, I_c \in\mathbb{R}^{HW}$ stand for the $c$-th column of $\hat{I}$ and $I$ respectively, then
$$
\mathrm{SCC}(\hat{I}, I) = \frac 1C\sum_{c=1}^C\frac{\sigma_{\mathrm{cov}_c}}{\hat\sigma_c\sigma_c}
$$

1. Q-index. $\hat{I}_c, I_c \in\mathbb{R}^{H\times W}$ is obtained by reshaping the $c$-th column of $\hat{I}$ and $I$ respectively, and they are divided into $B$ patches, denoted by $\hat{p}_c^{(i)}, p_c^{(i)}, i\in\{1, 2, \cdots, B\}$ (with uniform window)
$$
\mathrm{Q}(\hat{I}, I) = \frac 1C\sum_{c=1}^C\frac 1B \sum_{i=1}^{B}\frac{2\hat\mu_c^{(i)}\mu_c^{(i)}}{{\hat\mu_c^{(i)}}^2 + {\mu_c^{(i)}}^2}\frac{2\hat\sigma_c^{(i)}\sigma_c^{(i)}}{{\hat\sigma_c^{(i)}}^2 + {\sigma_c^{(i)}}^2}\frac{\sigma_{\mathrm{cov}_c}^{(i)}}{\hat\sigma_c^{(i)}\sigma_c^{(i)}}
$$

1. SSIM. $\hat{I}_c, I_c \in\mathbb{R}^{H\times W}$ is obtained by reshaping the $c$-th column of $\hat{I}$ and $I$ respectively, and they are divided into $B$ patches, denoted by $\hat{p}_c^{(i)}, p_c^{(i)}, i\in\{1, 2, \cdots, B\}$
then
$$
\mathrm{SSIM}(\hat{I}, I) = \frac 1C\sum_{c=1}^C \frac 1B \sum_{i=1}^{B}\mathrm{Luminance}(\hat{p}_c^{(i)}, p_c^{(i)})\mathrm{Contrast}(\hat{p}_c^{(i)}, p_c^{(i)})\mathrm{Structure}s(\hat{p}_c^{(i)}, p_c^{(i)})\,,
$$
where (with Gaussian window)
$$
\mathrm{Luminance}(\hat{p}_c^{(i)}, p_c^{(i)}) = \frac{2\hat{\mu}_c^{(i)}\mu_c^{(i)} + c_1}{{\hat{\mu}_c^{(i)}}^2 + {\mu_c^{(i)}}^2 + c_1}\,,
$$
$$
\mathrm{Contrast}(\hat{p}_c^{(i)}, p_c^{(i)}) = \frac{2\hat{\sigma}_c^{(i)}\sigma_c^{(i)} + c_2}{{\hat{\sigma}_c^{(i)}}^2 + {\sigma_c^{(i)}}^2 + c_2}\,,
$$
$$
\mathrm{Structure}(\hat{p}_c^{(i)}, p_c^{(i)}) = \frac{{\sigma}_{\mathrm{cov}_c}^{(i)} + c_3}{{\hat{\sigma}_c^{(i)}}\sigma_c^{(i)} + c_3}\,,
$$
$c_1 = (0.01\mathrm{MAX})^2, c_2 = (0.03\mathrm{MAX})^2, c_3 = c_2/2$, and $\mathrm{MAX}$ is the dynamic range of the pixel-values (typically, $2^8 - 1$).

1. ERGAS. Denote the ratio of spatial resolution of PAN and LRMS by $\mathrm{scale}$ (e.g. $2$ or $4$), then
$$
\mathrm{ERGAS}(\hat{I}, I) = 100\mathrm{scale}\sqrt{\frac 1C\sum_{c=1}^C\left(\frac{\mathrm{RMSE}_c}{\mu_c}\right)^2}\,,
$$
where $\mathrm{RMSE}_c$ is the RMSE between $\hat{I}_c$ and $I_c$; $\mu_c$ is mean of $I_c$.

1. QNR. The index is based on Q-index. Denote LM image with $C$ spectral bands as $I^\mathrm{(LM)}$, fused HM image with $C$ spectral bands as $\hat I$ and panchromatic image as $P$. And $P^\mathrm{(LM)}$ is the corresponding downsampled pancrhomatic image accoding to MTF of satellites.
    - Spectral Distortion Index.
    $$
    D_{\lambda} = \sqrt{\frac{2}{C(C-1)}\sum_{i=1}^{C}\sum_{j>i}^{C}|Q(\hat{I}_i, \hat{I}_j) - Q(I^\mathrm{(LM)}_i, I^\mathrm{(LM)}_j)|^p}^{p} \,.
    $$
    Usually, $p = 1$.
    - Spatial Distortion Index.
    $$
    D_s = \sqrt{\frac{1}{C}\sum_{i=1}^{C}|Q(\hat{I}_i, P) - Q(I^\mathrm{(LM)}_i,P^\mathrm{(LM)})|^q}^{q} \,.
    $$
    Usually, $q = 1$
    - QNR.
    $$
    \mathrm{QNR}(\hat I, I^\mathrm{(LM)}, P) = (1 - D_\lambda)^\alpha(1 - D_s)^\beta \,.
    $$
    Usually, $\alpha = \beta = 1$.
